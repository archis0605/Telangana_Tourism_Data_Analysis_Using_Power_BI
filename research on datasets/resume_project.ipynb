### Concatenate Domestic Data from each year into one CSV

```python
files1 = [file for file in os.listdir(r"C:/Users/archi/Kaggle_Competition/Telengana Tourism Dataset/domestic_visitors")]

domestic_2016_2019 = pd.DataFrame()

for file in files1:
    df1 = pd.read_csv(r"C:/Users/archi/Kaggle_Competition/Telengana Tourism Dataset/domestic_visitors/"+file)
    domestic_2016_2019 = pd.concat([domestic_2016_2019, df1])
    
domestic_2016_2019.to_csv("dvisitiors_2016_2019.csv", index=False)
```

### Concatenate Foreign Data from each year into one CSV

```python
files2 = [file for file in os.listdir(r"C:/Users/archi/Kaggle_Competition/Telengana Tourism Dataset/foreign_visitors")]

foreign_2016_2019 = pd.DataFrame()

for file in files2:
    df2 = pd.read_csv(r"C:/Users/archi/Kaggle_Competition/Telengana Tourism Dataset/foreign_visitors/"+file)
    foreign_2016_2019 = pd.concat([foreign_2016_2019, df2])
    
foreign_2016_2019.to_csv("fvisitiors_2016_2019.csv", index=False)
```
